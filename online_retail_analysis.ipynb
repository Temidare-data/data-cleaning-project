# Loading the Dataset

## The dataset is loaded using pandas for analysis.

```python
df = pd.read_excel("Online Retail.xlsx")
```

```python
df
```

# Data Overview

## Understanding the structure of the dataset.

```python
df.info()
```

# Data Cleaning

## Handling missing values, duplicates, and invalid data.

```python
df = df.dropna(subset=['CustomerID'])
df = df.drop_duplicates()
df = df[df['Quantity'] > 0]
```

```python
df.info()
```

# Feature Engineering

## Creating a new column for total transaction value.

```python
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
```

```python
df.head(10)
```

# Data Analysis

## Exploring key insights from the dataset.

```python
df['Description'].value_counts().head(10)
df['Country'].value_counts().head(10)
df['TotalPrice'].sum()
```

# Data Visualization

## Visualizing top countries by number of orders.

```python
df['Country'].value_counts().head(5).plot(kind='bar')
plt.xticks(rotation=45)
plt.title("Top 5 Countries by Orders")
plt.show()
```

# Conclusion

## The United Kingdom had the highest number of orders
